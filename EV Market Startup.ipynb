```python
import pandas as pd
import numpy as np
```

```python
df_loc=pd.read_csv("EV_India.csv")
```

```python
df_loc
```

```python
df_loc=df_loc.drop(['Sr. No.','Total Non-Electric Vehicle','Total'],axis=1)
```

```python
df_loc
```

```python
n=df_loc.iloc[:,-1].values
```

```python
print(n)
```

```python
import math
dff=[]
for i in n:
    z=0
    n1=[]
    if(i=='nan'):
        n1.append(0)
    elif(type(i)==float):
        n1.append(0)
    else:
        for j in i:
            if(j!=','):
                n1.append(j)
        l=len(n1)-1
        for k in n1:
            z+=int(k)*math.pow(10,l)
            l=l-1
    dff.append(z)
```

```python
dff
```

```python
df_loc['Total Electric Vehicle']=dff
```

```python
df_loc
```

```python
df_loc.sort_values(by=['Total Electric Vehicle'], ascending=False)
```

```python
from matplotlib import pyplot as plt
plt1=plt.figure(figsize=(15,15))
plt.bar(df_loc.iloc[:,0],df_loc.iloc[:,1])
plt.xticks(rotation =90)
plt.show()
```

From the above dataframe we can say that the UttarPradesh State has more number of Electric Vehicles followed by Delhi,Karnataka,Maharastra.Starting an EV startup from here will be more profitable compared to other states.

```python
df=pd.read_csv('Salary_Data.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Gender']=le.fit_transform(df['Gender'])
df['Education Level']=le.fit_transform(df['Education Level'])
```

```python
df.head()
```

```python
df=df.drop(['Job Title','Years of Experience'],axis=1)
```

```python
df.head()
```

```python
df.columns
```

```python
col=['Age', 'Gender', 'Education Level', 'Salary']
```

```python
df.describe()
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
sc_x=sc.fit_transform(df.iloc[:,[0,2,3]])
```

```python
print(sc_x)
```

```python
df['Age']=sc_x[:,0]
```

```python
df['Education Level']=sc_x[:,1]
```

```python
df['Salary']=sc_x[:,2]
```

```python
df.head()
```

```python
from sklearn.decomposition import PCA
pca=PCA()
df_pca=pca.fit_transform(df)
```

```python
df_pca
```

```python
c_pca=['PC1','PC2','PC3','PC4']
variance=pca.explained_variance_ratio_
sd=np.sqrt(pca.explained_variance_)
c_v=np.cumsum(pca.explained_variance_ratio_)
print("Variance")
print(variance)
print()
print("Standard Deviation")
print(sd)
print()
print("Cumulative Variance")
print(c_v)
```

```python
pos=pd.DataFrame(sd,index=c_pca,columns=['Standard Deviation'])
```

```python
pos
```

```python
pos['Variance']=variance
pos['Cumulative Variance']=c_v
```

```python
pos
#From this we can say that PC1 and PC2 combinely has 85% of information 
```

```python
r=pca.components_.T
```

```python
pos1=pd.DataFrame(r,index=col,columns=c_pca)
```

```python
pos1
```

```python
plt1=plt.figure(figsize=(10,10))
plt.scatter(df_pca[:, 0], df_pca[:, 1], color='grey',s=8)
for i, label in enumerate(col):
    plt.annotate(label, (r[i, 0], r[i, 1]),color='blue',size=10)
    plt.arrow(0, 0, r[i, 0], r[i, 1],head_width=0.05, head_length=0.05,color='red')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('PCA Projection')
plt.show()
```

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
    model=KMeans(n_clusters=i)
    model.fit(df)
    wcss.append(model.inertia_)
```
